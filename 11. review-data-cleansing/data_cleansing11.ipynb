```python
import pandas as pd
import numpy as np

data = pd.read_excel('accountMessage.xlsx') #首先读取添加好标注的数据集
data
```

```python
#接着按“完全合一”的准则先来处理--完整性，这里可发现age和weight以及f0612有缺失值；第9行整行存在空值
#这是因为此数据集很小所以容易一眼看出来，当数据集很大的时候就要用下面的方法来检测缺失值：
data.isnull() 
```

```python
#奇怪，官方文档里面用的是isna()，但是掌柜这里报错。。。不知道是不是版本的问题
data.isna()
```

```python
#先处理缺失值--删除法
#这里用的是dropna的all方法，指的是删除整列/整行的缺失值;如果用any就会把只要存在nan值的行/列删除掉
data.dropna(how='all',inplace=True)
```

```python
data  #再次查看删除后的数据
```

```python
#接着对部分数据进行填充,主要用到fillna（）方法
#先用平均值对年龄进行填充
data['Age'].fillna(data['Age'].mean(),inplace=True)
data
```

```python
#再用众位数来填充体重
weight_max = data['Weight'].value_counts().index[0]
data['Weight'].fillna(weight_max, inplace=True)
data
```

```python
#用相邻值的后值填充来填充女性三围的数据缺失值
data['f0612'].fillna(method='bfill',inplace=True)  
data
```

```python
#接下来是数据的全面性
data['Age'] = data['Age'].round(1)  #统一年龄的小数点位数，保持为一开始的只有一位
data
```

```python
#接着统一体重的单位为kgs 
data['Weight'] = data['Weight'].str.replace('lbs','kgs') #陈老师的解法用了循环，掌柜觉得复杂了，直接用str.replace()方法即可
data
```

```python
#下面进行继续合法性：这里需要处理的是非ASCII字符，删除或者替换即可
#这里是用正则来删除非ASCII字符
data['Name'].replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
data
```

```python
#接下来用替换方法，替换非ASCII字符
#翻遍全网，终于在一个细小的StackOverflow里面找到一个同样问题的朋友😂
repl_dict = {'é':'e', 'öö':'oo'}
for k,v in repl_dict.items():
    data.loc[data.Name.str.contains(k), 'Name'] = data.Name.str.replace(pat=k,repl=v)
    
data
```

```python
#最后再看唯一性：删除一对多以及重复的数据
data[['First Name','Last Name']] = data['Name'].str.split(expand=True) #添加expand参数=True表示返回一个数据框
data.drop('Name',axis=1,inplace=True)
```

```python
data
```

```python
#最后调整一下列的顺序
data = data[['Number','First Name','Last Name','Age','Weight','m0006','m0612','m1218','f0006','f0612','f1218']]
data
```

```python
#最后去重：
data.duplicated(['First Name','Last Name','Age','Weight','m0006','m0612','m1218','f0006','f0612','f1218']) #可以先查询哪些列有重复数据
```

```python
df = data.drop_duplicates('First Name')
```

```python
df
```

```python
#最后的最后把干净的数据写入Excel表格
df.to_excel('accountMessage_clean.xlsx', sheet_name='Sheet1')
```

```python
#这是用随机森林拟合预测后来填充缺失的Age值
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor


def set_missing_ages(df):

    # 把已有的数值型特征取出来丢进Random Forest Regressor中
    age_df = df[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]

    # 乘客分成已知年龄和未知年龄两部分
    known_age = age_df[age_df.Age.notnull()].as_matrix()
    unknown_age = age_df[age_df.Age.isnull()].as_matrix()

    # y即目标年龄
    y = known_age[:, 0]

    # X即特征属性值
    X = known_age[:, 1:]

    # fit到RandomForestRegressor之中
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
    rfr.fit(X, y)

    # 用得到的模型进行未知年龄结果预测
    predictedAges = rfr.predict(unknown_age[:, 1:])
    # print predictedAges
    # 用得到的预测结果填补原缺失数据
    df.loc[ (df.Age.isnull()), 'Age' ] = predictedAges 

    return df, rfr

if __name__ == '__main__':
    df = pd.read_csv('train.csv')
    set_missing_ages(df)
```

```python
#这是用随机森林拟合预测后来填充缺失的Age值
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

df = pd.read_csv('train.csv')
age_df = df[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]
print(age_df)
```

```python
# 再把乘客按年龄分成无缺失值部分和有缺失值部分
age_not_null = age_df[age_df.Age.notnull()].as_matrix()
age_null = age_df[age_df.Age.isnull()].as_matrix()

# X即特征属性值
X = age_not_null[:, 1:]
# y即目标年龄
y = age_not_null[:, 0]

#用随机森林进行拟合
rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
rfr.fit(X, y)

# 用得到的模型对有缺失值的年龄进行预测
predicted_age = rfr.predict(age_null[:, 1:])
# print predicted_age
# 用得到的预测结果填补原缺失数据
df.loc[ (df.Age.isnull()), 'Age' ] = predicted_age

df
```
